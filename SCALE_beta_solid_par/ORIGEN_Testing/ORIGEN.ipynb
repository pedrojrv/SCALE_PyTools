```python
import os 
from natsort import natsorted
import pandas as pd
```

```python
csv_directory = "." 

print("Searching directory for .csv files...")
names = []
for root, dirs, files in os.walk(csv_directory):
    for file in files:
        if file.endswith("s.csv"):
            #print(os.path.join(root, file))
            names.append(os.path.join(root, file))

if not names:
    print("List is empty")

names = natsorted(names)
names
```

```python
dfs = []

for file in names:
    df1 = pd.read_csv(file).T
    df1.columns = df1.iloc[0]
    df1 = df1.drop(df1.index[0])
    df1.columns = df1.columns.str.replace(' ', '')
    df1 = df1.drop(labels=["step"], axis=1)
    df1 = df1.reset_index(drop=True)
    dfs.append(df1)
    
for x in ["3:tc099m", "3:tc099", "3:mo099"]:  
    ax = dfs[0].plot(x="time", y=x, figsize=(14,8))
    for i in range (1,5):
        dfs[i].plot(x="time", y=x, ax=ax)
    ax.legend(["50", "100", "200", "400", "730"])
    ax.set_title(x + " Concentration in Fuel Salt")
    # ax.set_xscale('log')

    array = []
    for i in range(len(dfs)):
        last = dfs[i][x].values[-1]
        array.append(last)
    
    print(array)

    for i in array:
        error = (i - array[-1])/array[-1]
        print(error)
```

```python
csv_directory = "." 

print("Searching directory for .csv files...")
names = []
for root, dirs, files in os.walk(csv_directory):
    for file in files:
        if file.endswith(".csv"):
            #print(os.path.join(root, file))
            names.append(os.path.join(root, file))

if not names:
    print("List is empty")

names = natsorted(names)
names.pop(2)
names
```

```python
dfs = []
for file in names:
    df1 = pd.read_csv(file).T
    df1.columns = df1.iloc[0]
    df1 = df1.drop(df1.index[0])
    df1.columns = df1.columns.str.replace(' ', '')
    df1 = df1.drop(labels=["step"], axis=1)
    df1 = df1.reset_index(drop=True)
    dfs.append(df1)
```

```python
steps = ["50", "100", "200", "400", "730"]
  
u = 0
z = 0
for i in range(0, 5):
    for x in ["3:tc099m", "3:tc099", "3:mo099"]:  
        ax = dfs[z].plot(x="time", y=x, figsize=(14,8))
        dfs[z+1].plot(x="time", y=x, ax=ax)
        ax.legend(["ALL", "Tc99-Mo99"])
        ax.set_title(x + " Concentration in Fuel Salt on " +  steps[i])
        # ax.set_xscale('log')
    z = z + 2
```
