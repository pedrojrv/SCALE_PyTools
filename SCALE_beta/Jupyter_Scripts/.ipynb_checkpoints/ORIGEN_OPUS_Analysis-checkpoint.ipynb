# Importing Modules

```python
import pandas as pd
import numpy as np
from decimal import Decimal
import matplotlib.pyplot as plt
```

# Setting path for analysis

```python
path = "./Apollo/MSDR/MSDR_beta.000000000000000000.plt"
path2 = "./Apollo/MSDR/MSDR_beta.000000000000000001.plt"
```

```python
df = pd.read_csv(path, sep="\s+", skiprows=5)
df2 = pd.read_csv(path2, sep="\s+", skiprows=5)
```

```python
df_testing = df.copy()
df_testing2 = df2.copy()
```

```python
df_testing = df_testing.transpose()
df_testing2 = df_testing2.transpose()
```

```python
df_testing.reset_index(level=0, inplace=True)
df_testing2.reset_index(level=0, inplace=True)
```

```python
df_testing = df_testing.rename(columns = {'index':'time'})
df_testing2 = df_testing2.rename(columns = {'index':'time'})
```

```python
df_testing.iloc[:,1:] = df_testing.iloc[:,1:].mul(62.1)
df_testing2.iloc[:,1:] = df_testing2.iloc[:,1:].mul(62.1)
```

```python
ax = df_testing2.plot(x="time", y="xe135", figsize=(10,8), title="Concentration vs Time")
ax.set_xlabel("Time (days)")
ax.set_ylabel("U-238 (g)")
```

```python
isotopes = list(df_testing2.columns)
isotopes.pop(0)
num_isotopes = len(isotopes)
print("There are {} isotopes in Mixture 7.".format(num_isotopes))
```

```python
# ax = df.iloc[0:252, [0,2]].plot('Group No', 'Critical Flux', figsize=(14, 12), legend=None, ylim=(0, 0.010))
# plt.gca().invert_xaxis()
for i in range(1, num_isotopes):    
    isotope = isotopes[i]

    fig, ax1 = plt.subplots(figsize=(10,8))

    color = 'tab:red'
    ax1.set_xlabel('time (s)')
    ax1.set_ylabel('Mixture 1 (grams)', color=color)
    ax1.set_title(isotope)
    ax1.plot(df_testing['time'], df_testing[isotope], color=color)
    ax1.tick_params(axis='y', labelcolor=color)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

    color = 'tab:blue'
    ax2.set_ylabel('Mixture 7 (grams)', color=color)  # we already handled the x-label with ax1
    ax2.plot(df_testing2['time'], df_testing2[isotope], color=color)
    ax2.tick_params(axis='y', labelcolor=color)

    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.savefig("./Images/" + isotope + ".png")
    plt.close()
```
